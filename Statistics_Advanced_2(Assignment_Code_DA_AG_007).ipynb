**Q1. What is hypothesis testing in statistics?**  
**Answer:**  
Hypothesis testing is a statistical procedure for making decisions about population parameters using sample data. It involves formulating a **null hypothesis (Hâ‚€)** and an **alternative hypothesis (Hâ‚)**, computing a **test statistic**, obtaining a **p-value** (or comparing to a critical value), and deciding to **reject** or **fail to reject** Hâ‚€ based on a chosen **significance level (Î±)**.

---

**Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
**Answer:**  
- **Null Hypothesis (Hâ‚€):** The baseline claimâ€”usually â€œno effectâ€ or â€œno differenceâ€ (e.g., population mean Î¼ = 50).  
- **Alternative Hypothesis (Hâ‚):** The competing claim indicating an effect or difference (e.g., Î¼ â‰  50, Î¼ > 50, or Î¼ < 50).  
**Difference:** Hâ‚€ is assumed true unless data provide strong evidence for Hâ‚. We test the evidence to possibly reject Hâ‚€ in favor of Hâ‚.

---

**Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**  
**Answer:**  
The **significance level (Î±)** is the probability threshold for rejecting Hâ‚€. Common values are **0.05 (5%)** or **0.01 (1%)**.  
- If **p-value â‰¤ Î±**, reject Hâ‚€ (evidence against it).  
- If **p-value > Î±**, fail to reject Hâ‚€.  
Thus, Î± represents the risk of making a **Type I error** (rejecting a true Hâ‚€).

---

**Q4. What are Type I and Type II errors? Give examples of each.**  
**Answer:**  
- **Type I Error (False Positive):** Rejecting Hâ‚€ when it is true.  
  - Example: Concluding a medicine is effective when it is not.  
- **Type II Error (False Negative):** Failing to reject Hâ‚€ when it is false.  
  - Example: Concluding a medicine is not effective when it actually works.  

---

**Q5. What is the difference between a Z-test and a T-test? Explain when to use each.**  
**Answer:**  
- **Z-test:** Used when population variance is known **or** when sample size is large (n > 30). Test statistic follows the **standard normal (Z)** distribution.  
- **T-test:** Used when population variance is **unknown** and sample size is **small** (n â‰¤ 30). Test statistic follows the **t-distribution** (heavier tails).  

**Rule of thumb:** Use a T-test for small samples with unknown variance, otherwise use a Z-test.

```python
# Q7. Implement hypothesis testing using Z-statistics for a sample dataset in Python.
# Show the Python code and interpret the results.
# (Include your Python code and output in the code box below.)

import numpy as np
from scipy import stats

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypotheses
# H0: mu = 50
# H1: mu â‰  50
mu_0 = 50
x_bar = np.mean(sample_data)
s = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (x_bar - mu_0) / (s / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", x_bar)
print("Z-statistic:", z)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject H0: Significant evidence against the null hypothesis.")
else:
    print("Fail to reject H0: No significant evidence against the null hypothesis.")
```

```python
# Q8. Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# (Include your Python code and output in the code box below.)

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)

# Confidence Interval (95%)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=sem)

# Plot
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean={mean:.2f}")
plt.axvline(ci[0], color='green', linestyle='dashed', label=f"95% CI Lower={ci[0]:.2f}")
plt.axvline(ci[1], color='green', linestyle='dashed', label=f"95% CI Upper={ci[1]:.2f}")
plt.legend()
plt.title("Normal Distribution with 95% CI")
plt.show()

print("Mean:", mean)
print("95% Confidence Interval:", ci)
```

```python
# Q9. Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram.
# Explain what the Z-scores represent in terms of standard deviations from the mean.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt

# Z-score function
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Example dataset
data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

# Plot histogram of z-scores
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("Mean of Z-scores:", np.mean(z_scores))
print("Standard Deviation of Z-scores:", np.std(z_scores))
```
