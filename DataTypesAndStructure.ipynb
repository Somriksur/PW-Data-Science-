1) What are data structures, and why are they important?

-- Data structures are ways of organizing and storing data to perform operations efficiently. They are important because they help in optimizing performance, memory usage, and enable faster data processing.

2) Explain the difference between mutable and immutable data types with examples ?

--Mutable data types can be changed after creation.
Example: list → a = [1, 2]; a[0] = 5 → a becomes [5, 2].

Immutable data types cannot be changed once created.
Example: tuple → t = (1, 2); t[0] = 5 → Error.

3) What are the main differences between lists and tuples in Python ?

--Lists are mutable (can be changed), while tuples are immutable (cannot be changed).
Lists use [ ] brackets; tuples use ( ).
Example: list = [1, 2], tuple = (1, 2)

4) Describe how dictionaries store data ?

--Dictionaries store data as key-value pairs. Each key is unique and maps to a value. Data is stored using a hashing mechanism for fast lookup.
Example: {'name': 'Alice', 'age': 25}

5) Why might you use a set instead of a list in Python ?

--A set is used when you need to store unique elements and perform fast membership tests. Unlike lists, sets automatically remove duplicates and are faster for checking if an item exists.

6) What is a string in Python, and how is it different from a list ?

--A string is a sequence of characters, while a list is a sequence of any data types.
Strings are immutable, but lists are mutable.
Example: "hello" is a string, [1, 'a', 3] is a list.

7) How do tuples ensure data integrity in Python ?

--Tuples are immutable, meaning their values cannot be changed after creation. This prevents accidental modifications and helps maintain data integrity.

8) What is a hash table, and how does it relate to dictionaries in Python ?

--A hash table is a data structure that stores data in key-value pairs using a hash function for fast access.
Python dictionaries are implemented using hash tables for efficient lookups and updates.

9) Can lists contain different data types in Python ?

--Yes, lists in Python can contain different data types, such as integers, strings, and even other lists.
Example: [1, "hello", 3.5, [2, 4]]

10) Explain why strings are immutable in Python ?

--Strings are immutable in Python to ensure memory efficiency and data safety. Once created, they cannot be changed, which helps avoid accidental modifications and makes them hashable for use as dictionary keys.

11) What advantages do dictionaries offer over lists for certain tasks ?

--Dictionaries offer faster lookups using keys (O(1) time) and are ideal when data is key-value paired, unlike lists that require linear search (O(n)) for value-based access.

12) Describe a scenario where using a tuple would be preferable over a list ?

--Use a tuple when you need fixed, unchangeable data, such as storing coordinates (x, y), because it ensures the data won’t be accidentally modified.

13) How do sets handle duplicate values in Python ?

--Sets automatically remove duplicates. When duplicate elements are added, only one copy is kept, ensuring all elements in a set are unique.

14) How does the “in” keyword work differently for lists and dictionaries ?

--For lists, "in" checks if a value exists. For dictionaries, it checks if a key exists.
Example: 'a' in [1, 'a'] vs 'key' in {'key': 1}

15) Can you modify the elements of a tuple? Explain why or why not ?

--No, tuples are immutable, meaning their elements cannot be changed after creation. This ensures the data remains constant and protected.

16) What is a nested dictionary, and give an example of its use case ?

--A nested dictionary is a dictionary within another dictionary.
Example: students = {'John': {'age': 20, 'grade': 'A'}} – used to store structured data like student records.

17) Describe the time complexity of accessing elements in a dictionary ?

--Accessing elements in a dictionary has an average time complexity of O(1) due to the use of hash tables for fast key lookups.

18) In what situations are lists preferred over dictionaries?

--Lists are preferred when you need to store ordered data or sequential collections without key-based access, such as a list of names or numbers.

19) Why are dictionaries considered unordered, and how does that affect data retrieval ?

--Dictionaries were traditionally unordered (before Python 3.7), meaning key order was not guaranteed. Though order is preserved now, retrieval still depends on key access, not position.

20) Explain the difference between a list and a dictionary in terms of data retrieval ?

--In a list, data is accessed using index positions (e.g., list[0]). In a dictionary, data is retrieved using keys (e.g., dict['name']).
