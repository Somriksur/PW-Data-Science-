```python
#Theory
# 1. Create a 2D NumPy array and calculate the sum of each row
import numpy as np
import pandas as pd

df = pd.read_csv("/players.csv")
arr = df.select_dtypes(include=[np.number]).iloc[:5, :3].values  # take first 3 numeric columns and 5 rows
row_sums = arr.sum(axis=1)
print("2D Array:\n", arr)
print("Sum of each row:", row_sums)
```

```python
# 2. Find the mean of a specific column in a DataFrame
df = pd.read_csv("/Bank_churn.csv")
mean_val = df['CreditScore'].mean()
print("Mean Credit Score:", mean_val)
```

```python
# 3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

plt.scatter(df['CreditScore'], df['Balance'])
plt.xlabel("Credit Score")
plt.ylabel("Balance")
plt.title("Scatter: Credit Score vs Balance")
plt.show()
```

```python
# 4. Correlation matrix and heatmap using Seaborn
import seaborn as sns

corr = df[['CreditScore', 'Balance', 'EstimatedSalary']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly
import plotly.express as px

fig = px.bar(df, x='Geography', y='Balance', title='Average Balance by Geography', color='Geography')
fig.show()
```

```python
# 6. Add a new column based on an existing one
df['Balance_Squared'] = df['Balance'] ** 2
print(df[['Balance', 'Balance_Squared']].head())
```

```python
# 7. Perform element-wise multiplication of two NumPy arrays
arr1 = np.array([10, 20, 30])
arr2 = np.array([2, 3, 4])
print("Multiplication:", arr1 * arr2)
```

```python
# 8. Line plot with multiple lines using Matplotlib
x = range(10)
plt.plot(x, df['CreditScore'].head(10), label='Credit Score')
plt.plot(x, df['Balance'].head(10), label='Balance')
plt.legend()
plt.title("Line Plot: Credit Score and Balance")
plt.show()
```

```python
# 10. Histogram using Seaborn
sns.histplot(df['EstimatedSalary'], kde=True)
plt.title("Estimated Salary Distribution")
plt.show()
```
